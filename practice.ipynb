```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assume we have a DataFrame 'df' with a categorical column 'Category'
df1 = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'A', 'B', 'C', 'A', 'B', 'C', 'C', 'C']
})

# Create a countplot
sns.countplot(x='Category', data=df1)

# Show the plot
plt.show()
```
